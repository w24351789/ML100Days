```python
import numpy as np
```

```python
#1.生成一個等差數列，首數為0，尾數為20，公差為1的數列。
```

```python
a = np.arange(0, 21, step=1)
```

```python
a
```

```python
#2.呈上題，將以上數列取出偶數。
```

```python
b = []
for num in a:
   if num%2 == 0:
       b.append(num)
```

```python
b
```

```python
#3.呈1題，將數列取出3的倍數。
```

```python
c = []
for num in a:
    if num%3 == 0 and num!=0:
        c.append(num)
```
