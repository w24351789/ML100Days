```python
import numpy as np
```

```python
#1.將下列清單(list1)，轉成維度為(5X6)的array，順序按列填充。(hint:order="F")
array1 = np.array(range(30))
```

```python
a = array1.reshape((5, 6), order='F')
a
```

```python
#2.呈上題的array，找出被6除餘1的數的索引
```

```python
b = np.where((a%6) == 1)
```
