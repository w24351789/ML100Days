1. 運用上列array計算反矩陣，乘上原矩陣，並觀察是否為單位矩陣?
2. 運用上列array計算特徵值、特徵向量?
3. 運用上列array計算SVD?

```python
import numpy as np
```

```python
#1. 運用上列array計算反矩陣，乘上原矩陣，並觀察是否為單位矩陣?
array1 = np.array([[10, 8], [3, 5]])
try:
    array1_inv = np.linalg.inv(array1)
except np.linalg.LinAlgError:
    print('cannot inv')
ans_1 = np.dot(array1, array1_inv)

ans_1
```

```python
#2. 運用上列array計算特徵值、特徵向量?
e_vals, e_ecvs = np.linalg.eig(array1)
e_vals
```
